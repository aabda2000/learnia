```python
import numpy as np
import matplotlib.pyplot as plt

def drawArrow(A, B,iter):
    for i in range(iter -1):
        plt.arrow(A[i], B[i], A[i+1] - A[i], B[i+1] - B[i],head_width=0.03, length_includes_head=True)
    
def inject(r,xo,iter):
    fx=np.arange(0,iter,dtype=float)
    x=np.arange(0,iter,dtype=float)
    x[0]=xo
    for i in range(iter -1):
        z=-4.0*r*x[i]*x[i]+ 4*r*x[i]
        fx[i]=z
        x[i+1]=fx[i]
    fx[i+1]=-4.0*r*x[i+1]*x[i+1]+ 4*r*x[i+1]
    
    print(x)
    print(fx)
    return (x,fx)

def trinome(r,xo,iter):
    x = np.linspace(0,1)
    w = np.linspace(0,1)
    fx = -4*r*x**2 + 4*r*x
    v=w
    xinj,yinj=inject(r,xo,iter)
    # setting the axes at the centre
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    # plot the function
    drawArrow(xinj,yinj,iter)
    plt.plot(x,fx, 'r')
    plt.scatter(xinj, yinj)
    plt.plot(v,w, 'y')
    # show the plot
    plt.show()
def main():
    trinome(0.8,0.05,100)
   


if __name__ == '__main__':
    main()
```
